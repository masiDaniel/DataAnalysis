```python
#show top 3 preferred shops that have the largest number of unique products
df[(df.is_preferred==1)]['shopid'].value_counts()[:3]
```

```python
#show top 3 categories that have the largest number of unique cross border items
df[(df.cb_option==1)]['category'].value_counts()[:3]
```

```python
#find top 3 shopid with the highest revenue
revenue=df['price']*df['sold_count']
df['revenue']=revenue
df.groupby('shopid')['revenue'].sum().sort_values(ascending=False)[:3]
```

```python
#find number of products with more than 3 variations
df['item_variation'].value_counts()[df['item_variation'].value_counts()>3]
```

```python
#identify duplicated listings within each shop(uif listing a and b have exactly same product title,product detailed description,and price, both listing a and b are considered duplicated)
are_duplicated = df.duplicated('item_name')
df[are_duplicated]
```

```python
import numpy as np
np.arange(0,456577)
```

```python
#mark these listings with True otherwise False and store the marking results in a new column called is_duplicated
df['is_duplicated']=are_duplicated
df
```

```python
#find duplicate listings that has less than 2 sold count and store the results in a new excel file named "duplicated_listings.xlsx"
filtered_df = df[(df.is_duplicated==True) & (df.sold_count<2)]
filtered_df.to_excel("duplicated_listings.xlsx", index=False)
```
